```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_init_cond(starter,length):
    num_str=str(starter).split('.')[-1]
    n=len(num_str)
    while n<length:
        num_str+=str(np.random.random()).split('.')[-1]
        n=len(num_str)
        if n>=length:
            num_str=num_str[0:n]
    return num_str
```

```python
start=float('0.'+generate_init_cond(0.1,10))

system_1=[int(char) for char in generate_init_cond(start,200)]
system_2=[int(char) for char in generate_init_cond(start,200)]

colors = plt.cm.cividis(np.linspace(0, 1, 3))


plt.rcParams["figure.figsize"] = (20,3)

plt.plot(range(len(system_1)),system_1,color=colors[0])
plt.plot(range(len(system_2)),system_2,color=colors[1])
plt.title('E_10')
plt.savefig('chaos_E10.png')
```
